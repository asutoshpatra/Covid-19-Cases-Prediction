```python
#Linear regression
from sklearn.model_selection import train_test_split
```

```python
#Tamil Nadu
tn = df[df.state == 'Tamil Nadu']
```

```python
tn
```

```python
#converting date-time to ordinal
tn['date']=tn['date'].map(dt.datetime.toordinal)
```

```python
#getting dependent variable and inpedent variable
x=tn[['date']]
y=tn['confirmed']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
print("x_test",x_test)
predicted_y = lr.predict(x_test)# test the model
print("predicted y records:",predicted_y)
print("real answers",y_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from math import sqrt
print("RMSE is:-")

print('Mean Absolute Error(MAE):', mean_absolute_error(y_test, predicted_y))
print('Mean Squared Error(MSE):', mean_squared_error(y_test, predicted_y))
print('Root Mean Squared Error(RMSE):', np.sqrt(mean_squared_error(y_test, predicted_y)))
```

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test, predicted_y, color = 'green')
plt.title ('Results ')
plt.xlabel('Date')
plt.ylabel('confirm cases')
plt.show()
```

```python
#explain equation now y=mx+c
m=lr.coef_  # slope
c=lr.intercept_   #intersecpt
print(" slope:",m) #slope
print("intercept",c)  

y=m*737976+c
print("Confirmed Cases(using y=mx+c)",y)


lr_result=lr.predict([[737976]])
print("Confirmed Cases(Using linear regression)",lr_result)
```
